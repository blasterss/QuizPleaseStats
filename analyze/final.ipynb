```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel("../main_data/data_classic.xlsx")
df1.head()
```

```python
import datetime


classic_dates = ['29.08.2024', '30.05.2024', '29.02.2024', '30.11.2023']
date_season = classic_dates[1]
date_season2 = classic_dates[2]
date_now = datetime.datetime.now().date().strftime('%d.%m.%Y')

start_date = pd.to_datetime(date_season2, format='%d.%m.%Y')
print(start_date)
end_date = pd.to_datetime(date_season, format='%d.%m.%Y')
df1['Дата'] = pd.to_datetime(df1['Дата'], format='%d.%m.%Y')
df1 = df1[(df1['Дата'] >= start_date) & (df1['Дата'] < end_date)] 
df1 = df1[df1['Название игры'].str.contains("Квиз, плиз! SPB", na=False)]
df1.head()
```

```python
df2 = pd.read_excel("finals.xlsx")
df2['Дата'] = pd.to_datetime(df2['Дата'], format='%d.%m.%Y')
df2 = df2[(df2['Дата'] == end_date)] 
df2.head()
```

```python
new_df2 = df2.drop(['Название игры', 'Ресторан', 'Дата', 'День недели', 'Номер пакета'], axis=1)
new_df1 = df1.drop(['Название игры', 'Ресторан', 'Дата', 'День недели', 'Номер пакета', 'Последний раунд'], axis=1)
```

```python
new_df2.head()
```

```python
df1_grouped = new_df1.groupby('Название команды').agg(
    Среднее_место=('Место', lambda x: round(x.mean(), 2)),
    Средний_балл=('Всего баллов', lambda x: round(x.mean(), 2)),
    Количество_игр=('Название команды', 'size')
).reset_index()
```

```python
result = pd.merge(new_df2, df1_grouped, on='Название команды', how='left')
result = result.rename({'Всего баллов': 'Всего баллов в финале', 'Место': 'Место в финале', 'Среднее_место': 'Среднее место в сезоне', 'Средний_балл': 'Средний балл в сезоне', 'Количество_игр': 'Кол-во игр в сезоне'}, axis=1)
result = result[['Название команды', 'Кол-во игр в сезоне', 'Среднее место в сезоне', 'Место в финале', 'Средний балл в сезоне', 'Всего баллов в финале']]
result.head(60)
```
