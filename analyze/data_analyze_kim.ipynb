```python
df = pd.concat([df1,df2])
df.head()
```

```python
data_dict = {}

# Получение уникальных названий команд
unique_teams = df1['Название команды'].unique()

# Итерация по уникальным командам
for team in unique_teams:
    # Фильтрация данных для текущей команды
    team_data = df1[df1['Название команды'] == team].copy()
    # Удаление столбца с названием команды, чтобы он не дублировался внутри каждого подфрейма
    team_data.drop(columns=['Название команды'], inplace=True)
    # Сохранение подфрейма в словаре, используя название команды в качестве ключа
    data_dict[team] = team_data
```

```python
games_per_restaurant = df.pivot_table(index='Название команды', columns='Ресторан', aggfunc='size', fill_value=0)

# Подсчет количества игр в каждом ресторане, где место было больше 6
games_per_restaurant_over_6 = df[df['Место'] < 6].pivot_table(index='Название команды', columns='Ресторан', aggfunc='size', fill_value=0)

# Подсчет количества игр в каждый день недели для каждого ресторана
games_per_weekday_per_restaurant = df.pivot_table(index='Название команды', columns='День недели', aggfunc='size', fill_value=0)

# Подсчет количества игр в каждый день недели, где место было больше 6, для каждого ресторана
games_per_weekday_over_6_per_restaurant = df[df['Место'] < 6].pivot_table(index='Название команды', columns='День недели', aggfunc='size', fill_value=0)

games_per_package = df.pivot_table(index='Название команды', columns='Номер пакета', aggfunc='size', fill_value=0)

# Подсчет количества игр в каждый день недели, где место было больше 6, для каждого ресторана
games_per_package_over_6 = df[df['Место'] == 1].pivot_table(index='Название команды', columns='Номер пакета', aggfunc='size', fill_value=0)

# Объединение всех таблиц
final_df = pd.concat([games_per_restaurant, games_per_restaurant_over_6, games_per_weekday_per_restaurant, games_per_weekday_over_6_per_restaurant, games_per_package, games_per_package_over_6], axis=1)
```
